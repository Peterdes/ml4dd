```python
def generate_ecfp(smiles, radius=3, nBits=2048, use_features=False, use_chirality=True):
    molecule = Chem.MolFromSmiles(smiles)
    if molecule is None:
        return np.zeros(nBits)
    feature_list = AllChem.GetMorganFingerprintAsBitVect(
        molecule,
        radius=radius,
        nBits=nBits,
        useFeatures=use_features,
        useChirality=use_chirality,
    )
    return np.array(feature_list)
```

```python
def generate_maccs(smiles):
    molecule = Chem.MolFromSmiles(smiles)
    if molecule is None:
        return np.zeros(167)
    maccs_keys = MACCSkeys.GenMACCSKeys(molecule)
    return np.array(maccs_keys)
```

```python
def calculate_descriptors_for_row(row):
    smiles = row['smiles']
    mol = Chem.MolFromSmiles(smiles)
    if mol is not None:
        # Calculate 2D descriptors
        descriptor_dict = {
            'SMILES': smiles,
            'MolWt': Descriptors.MolWt(mol),
            'LogP': Descriptors.MolLogP(mol),
            'NumHAcceptors': Descriptors.NumHAcceptors(mol),
            'NumHDonors': Descriptors.NumHDonors(mol),
            'NumHeteroatoms': Descriptors.NumHeteroatoms(mol),
            'NumRotatableBonds': Descriptors.NumRotatableBonds(mol),
            'NumAliphaticCarbocycles': Descriptors.NumAliphaticCarbocycles(mol),
            'NumAliphaticHeterocycles': Descriptors.NumAliphaticHeterocycles(mol),
            'NumAliphaticRings': Descriptors.NumAliphaticRings(mol),
            'NumAromaticCarbocycles': Descriptors.NumAromaticCarbocycles(mol),
            'NumAromaticHeterocycles': Descriptors.NumAromaticHeterocycles(mol),
            'NumAromaticRings': Descriptors.NumAromaticRings(mol),
            'RingCount': Descriptors.RingCount(mol),
            'NHOHCount': Descriptors.NHOHCount(mol),
            'FractionCSP3': Descriptors.FractionCSP3(mol),
            'MinPartialCharge': Descriptors.MinPartialCharge(mol),
            'MaxPartialCharge': Descriptors.MaxPartialCharge(mol),
            'TPSA': Descriptors.TPSA(mol),
        }

        # Calculate 3D descriptors
        mol_3d = Chem.AddHs(mol)
        Chem.AllChem.EmbedMolecule(mol_3d, randomSeed=42)
        if mol_3d.GetNumConformers() > 0:
            descriptor_dict.update({
                'NPR1': Descriptors3D.NPR1(mol_3d),
                'NPR2': Descriptors3D.NPR2(mol_3d),
                'InertialShapeFactor': Descriptors3D.InertialShapeFactor(mol_3d),
                'RadiusOfGyration': Descriptors3D.RadiusOfGyration(mol_3d),
                'SpherocityIndex': Descriptors3D.SpherocityIndex(mol_3d),
            })
        else:
            print(f"Warning: 3D conformation generation failed for {smiles}")
            descriptor_dict.update({
                'NPR1': None,
                'NPR2': None,
                'InertialShapeFactor': None,
                'RadiusOfGyration': None,
                'SpherocityIndex': None,
            })

        # Calculate ECFP
        ecfp = generate_ecfp(smiles)
        ecfp_dict = {f'ECFP_{i}': ecfp[i] for i in range(len(ecfp))}
        descriptor_dict.update(ecfp_dict)

        # Calculate MACCS keys
        maccs = generate_maccs(smiles)
        maccs_dict = {f'MACCS_{i}': maccs[i] for i in range(len(maccs))}
        descriptor_dict.update(maccs_dict)

        # Preserve other columns
        for col in ['KIT_(T6701_mutant)', 'KIT_(V560G_mutant)', 'KIT',
                    'CLASS_KIT_(T6701_mutant)', 'CLASS_KIT_(V560G_mutant)', 'CLASS_KIT']:
            if col in row:
                descriptor_dict[col] = row[col]

        return pd.Series(descriptor_dict)
    else:
        print(f"Could not parse SMILES: {smiles}")
        return pd.Series({'SMILES': smiles})
```

```python
def inverse_sigmoid(x):
    return np.log(x / (1 - x))

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

kit_columns = ['KIT_(T6701_mutant)', 'KIT_(V560G_mutant)', 'KIT']

offset = 1
divby = 100
epsilon = 1e-2

# Initialize MinMaxScaler to scale values to (0, 1)
#scaler = MinMaxScaler(feature_range=(0.01, 0.99))

# Fit and transform the data using MinMaxScaler
#kit_vals_normalized_train = scaler.fit_transform(train_df[kit_columns])
#kit_vals_normalized_test = scaler.fit_transform(test_df[kit_columns])

kit_vals_normalized_train = (train_df[kit_columns] + offset) / divby
kit_vals_normalized_test = (test_df[kit_columns] + offset) / divby

kit_vals_normalized_train[kit_vals_normalized_train <= 0.] = epsilon
kit_vals_normalized_test[kit_vals_normalized_test <= 0.] = epsilon

kit_vals_normalized_train[kit_vals_normalized_train >= 1.] = 1. - epsilon
kit_vals_normalized_test[kit_vals_normalized_test >= 1.] = 1. - epsilon

# Apply the inverse sigmoid transformation
kit_vals_processed_train = inverse_sigmoid(kit_vals_normalized_train)
kit_vals_processed_test = inverse_sigmoid(kit_vals_normalized_test)

# Convert the processed values back to a DataFrame
kit_train_df = pd.DataFrame(kit_vals_processed_train, columns=kit_columns)
kit_test_df = pd.DataFrame(kit_vals_processed_test, columns=kit_columns)
```

```python
corr_matrix = train_df.drop(columns=['SMILES']).corr()

target_vars = ['KIT_(T6701_mutant)', 'KIT_(V560G_mutant)', 'KIT']

top_corr_dict = {}
for target in target_vars:
    # Extract correlations with the target variable
    target_correlations = corr_matrix[target].drop(target)  # Drop the correlation of target with itself

    # Sort correlations by absolute value to find the most significant correlations
    sorted_correlations = target_correlations.abs().sort_values(ascending=False)

    # Select the top N features
    top_n = 20
    top_features = sorted_correlations.head(top_n).index

    # Get the actual correlation values for these top features
    top_correlations = target_correlations[top_features]

    # Store the result in the dictionary
    top_corr_dict[target] = top_correlations
```

```python
for target, top_correlations in top_corr_dict.items():
    plt.figure(figsize=(10, 6))
    sns.barplot(x=top_correlations.index, y=top_correlations.values)
    plt.xticks(rotation=90)
    plt.xlabel('Features')
    plt.ylabel(f'Correlation with {target}')
    plt.title(f'Top {top_n} Feature Correlations with {target}')
    plt.show()
```

```python
# Define your target variables
target_variables = ['KIT_(T6701_mutant)', 'KIT_(V560G_mutant)', 'KIT']

# Define the prefixes to filter features
prefixes = ['MACCS_', 'ECFP_']

# Dictionary to hold the relevant correlations for each target variable
relevant_correlations_dict = {}

for target in target_variables:
    # Extract correlations with the target variable, excluding other targets
    target_correlations = corr_matrix[target].drop(target_variables)

    # Filter features based on the specified prefixes
    relevant_features = [col for col in target_correlations.index if col.startswith(tuple(prefixes))]

    # Get the correlation values for these relevant features
    relevant_correlations = target_correlations[relevant_features]

    # Store the result in the dictionary
    relevant_correlations_dict[target] = relevant_correlations
```

```python
for target, correlations in relevant_correlations_dict.items():
    plt.figure(figsize=(10, 6))
    sns.histplot(correlations, bins=20, kde=True)
    plt.xlabel('Correlation')
    plt.ylabel('Frequency')
    plt.title(f'Histogram of Correlations for {target} (Features starting with MACCS_ or ECFP_)')
    plt.show()
```

```python
# Set the correlation threshold
threshold = 0.1

# Set to store relevant features
relevant_features = set()

for target in target_variables:
    # Extract correlations with the target variable, excluding other targets
    target_correlations = corr_matrix[target].drop(target_variables)

    # Filter features based on the correlation threshold
    high_correlation_features = target_correlations.index[target_correlations.abs() >= threshold]

    # Add relevant features to the set
    relevant_features.update(high_correlation_features)

# Add the target variables back to the set of relevant features
relevant_features.update(target_variables)

# Convert the set of relevant features to a list
relevant_features = list(relevant_features)

# Filter the DataFrame to retain only the relevant features
filtered_df_train = train_df[relevant_features]
filtered_df_test = test_df[relevant_features]

# Display the shape of the original and filtered DataFrames
print(f'Original DataFrame shape: {train_df.shape}')
print(f'Filtered DataFrame shape: {filtered_df_train.shape}')

# Display the shape of the original and filtered DataFrames
print(f'\nOriginal DataFrame shape: {test_df.shape}')
print(f'Filtered DataFrame shape: {filtered_df_test.shape}')
```
