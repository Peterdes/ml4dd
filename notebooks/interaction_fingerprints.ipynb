## ProLIF

```python
from rdkit import Chem

from config import settings

protein_name = "KIT_wt"
protein_pdb_path = settings.data_path / protein_name / "protein.pdb"
protein_pdbqt_path = settings.data_path / protein_name / "protein.pdbqt"

ligand_sdf_path = settings.data_path / protein_name / "smina_outputs" / "poses_0.sdf"
ligand2_sdf_path = settings.data_path / protein_name / "smina_outputs" / "poses_1.sdf"
```

```python
from rdkit.DataStructs import TanimotoSimilarity
import prolif

protein_mol = Chem.MolFromPDBFile(str(protein_pdb_path))
protein_suppl = prolif.pdbqt_supplier(str(protein_pdbqt_path), protein_mol)
protein = prolif.molecule.Molecule(protein_mol)
ligand_suppl = prolif.sdf_supplier(str(ligand_sdf_path))
ligand = ligand_suppl[0]

prolif_fp_generator = prolif.Fingerprint(["HBDonor", "HBAcceptor", "PiStacking", "CationPi",
                         "Cationic"])

prolif_fp = prolif_fp_generator.generate(protein, ligand)
prolif_fp
```
